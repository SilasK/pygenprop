```python
def generate_fasta_table_rows(file_handle):
    for seq in skbio.io.read(fasta_handle, format='fasta'):
        yield seq.metadata['id'], str(seq)
```

```python
with open('/Users/lee/Desktop/rand_data/fastas/CA_L227_2013_22_1_ORFs.faa', 'r') as fasta_handle:
    fasta_frame = pd.DataFrame(data=generate_fasta_table_rows(fasta_handle), columns=['identifier','sequence'])
```

```python
with open('/Users/lee/Desktop/rand_data/Inter_pro_scan_results/CA_L227_2013_22_1.tsv', 'r') as inter_pro_handle:
    column_names = ['protien_accession','sequence_md5','sequence_length','analysis','signature_accession',
                    'signature_description','start_location','stop_location','score','status','date',
                    'interpro_accession','interpro_description','go_terms','pathway_terms']
    interpro_frame = pd.read_csv(inter_pro_handle, sep='\t', header=None, names=column_names,)
```

```python
def filter_tables_to_best_hits(fasta_frame, interpro_frame):
    filtered_interpro_frame = interpro_frame.loc[interpro_frame.groupby(['signature_accession'], as_index=False)['score'].idxmin()][['protien_accession', 'signature_accession','score']].sort_values(['protien_accession', 'signature_accession'])
    filtered_fasta_frame = fasta_frame[fasta_frame['identifier'].isin(filtered_interpro_frame['protien_accession'].drop_duplicates())]
    return filtered_fasta_frame, filtered_interpro_frame
```
